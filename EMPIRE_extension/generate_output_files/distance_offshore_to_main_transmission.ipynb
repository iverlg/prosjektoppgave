#### Calculate distance from offshore node all the way to onshore node (instead of to shore)

```python
import pandas as pd

offshore_nodes = pd.read_csv("../Output/offshore_to_onshore_nodes.csv")[["fromNode", "toNode", "dist"]]
offshore_nodes["unique_id"] = offshore_nodes.apply(lambda x: ''.join(sorted([x["fromNode"],x["toNode"]])),axis=1)
offshore_nodes
```

Create lookup table based on all nodes distances

```python
all_nodes_dist = pd.read_csv("../Output/dist_all_areas_scaled.csv")
all_nodes_dist["unique_id"] = all_nodes_dist.apply(lambda x: ''.join(sorted([x["Node"],x["to_Node"]])),axis=1)
nodes_lookup = all_nodes_dist.set_index(['unique_id'])['distance']
nodes_lookup
```

```python
offshore_nodes["dist"] = offshore_nodes\
                .apply(lambda row: nodes_lookup.get((row['unique_id']), row["dist"]), axis=1)
offshore_nodes = offshore_nodes.drop(columns=["unique_id"])
offshore_nodes
```

Export CSV

```python
offshore_nodes.to_csv("../Output/offshore_to_onshore_main_transmission.csv", index=False, encoding='utf-8-sig')
```
