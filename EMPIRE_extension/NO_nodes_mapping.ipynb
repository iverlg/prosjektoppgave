### Used to extend dict_countries and windfarmNodes in Public-EMPIRE run_HydrogenAllCountries.py

```python
import pandas as pd

df = pd.read_csv("Output/OW_nodes.csv")
df = df.iloc[12:]
df
```

```python
mapping = dict({
    "Nordvest": "NVE",
    "Vestavind": "VV",
    "Sønnavind": "SNV",
    "Sørvest": "SRV",
    "Nordavind": "NAV",
})

df["Node_mapping"] = df["Node"].apply(lambda n: ''.join([mapping.get(i, i) for i in n.split()]))
df
```

```python
dict_countries_extension = dict({})

for _, row in df.iterrows():
    dict_countries_extension[row["Node_mapping"]] = row["Node"]

dict_countries_extension
```
