### Correct Vestavind B from NO5 to NO3

Originally Vestavind B assigned to NO5, but to be consistent with NVE changed to NO3

```python
import pandas as pd
```

```python
def change_to_NO5(df_row, col1, col2):
    combo = ["Vestavind B", "NO5"]
    if df_row[col1] in combo and df_row[col2] in combo:
        if df_row[col1] == "NO5":
            df_row[col1] = "NO3"
        elif df_row[col2] == "NO5":
            df_row[col2] = "NO3"
    return df_row
```

Directional lines

```python
dl = pd.read_csv("../Output/directional_lines.csv")
col1 = "FromNode"
col2 = "ToNode"
wind_park = "Vestavind B"
dl[(dl[col1] == wind_park) | (dl[col2] == wind_park)]
```

```python
dl = dl.apply(lambda row: change_to_NO5(row, col1, col2), axis=1)
dl[(dl[col1] == wind_park) | (dl[col2] == wind_park)]
```

```python
dl.to_csv("../Output/directional_lines.csv", index=False, encoding="utf-8-sig")
```

Dist offshore areas

```python
file = "../Output/offshore_to_onshore_nodes_scaled.csv"
```

```python
def change_to_NO5(df_row, col1, col2):
    combo = ["Vestavind B", "NO5"]
    if df_row[col1] in combo and df_row[col2] in combo:
        if df_row[col1] == "NO5":
            df_row[col1] = "NO3"
        elif df_row[col2] == "NO5":
            df_row[col2] = "NO3"
        df_row["newOnshoreNode"] = "Vestavind B-NO3"
    return df_row
```

```python
df = pd.read_csv(file)
col1 = "fromNode"
col2 = "toNode"
wind_park = "Vestavind B"
df[(df[col1] == wind_park) | (df[col2] == wind_park)]
df = df.apply(lambda row: change_to_NO5(row, col1, col2), axis=1)
df[(df[col1] == wind_park) | (df[col2] == wind_park)]
```

```python
df.to_csv(file, index=False, encoding="utf-8-sig")
```
