#### Directional lines (transmission lines) between OW areas

```python
import pandas as pd 
import numpy as np
```

Offshore to offshore - keep only lines where distance is less than 500 km

Type of cable: HVDC_Cable

```python
offshr_to_offshr = pd.read_csv("Output/dist_offshore_areas_scaled.csv")
offshr_to_offshr = offshr_to_offshr[offshr_to_offshr["distance"] < 500]
offshr_to_offshr["LineType"] = "HVDC_Cable"
offshr_to_offshr
```

Offshore to onshore

LineType: HVDC_OffshoreToOnshore

```python
offshr_to_onshr = pd.read_csv("Output/offshore_to_onshore_nodes_scaled.csv")
offshr_to_onshr = offshr_to_onshr[["fromNode", "toNode", "dist"]]
offshr_to_onshr["LineType"] = "HVDC_OffshoreToOnshore"
offshr_to_onshr
```

Same format and combine

```python
offshr_to_offshr = offshr_to_offshr.rename(columns={"Node": "FromNode", "to_Node": "ToNode", "distance": "lineLength in km"})
offshr_to_onshr = offshr_to_onshr.rename(columns={"fromNode": "FromNode", "toNode": "ToNode", "dist": "lineLength in km"})
directional_lines = pd.concat([offshr_to_offshr, offshr_to_onshr], ignore_index=True).reset_index(drop=True)
directional_lines
```

Export CSV

```python
directional_lines.to_csv("Output/directional_lines.csv", index=False, encoding="utf-8-sig")
```
