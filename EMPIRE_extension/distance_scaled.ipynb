### Normalize/scale distances used in EMPIRE 

"normalized BE-NL is 175 km" --> Find scaling factor and adj other distances

```python
import pandas as pd 
import geopandas as gpd
```

```python
from shapely import wkt
df_nodes = pd.read_csv("Output/nodes.csv")
df_nodes["geometry"] = df_nodes["geometry"].apply(wkt.loads)
nodes = gpd.GeoDataFrame(df_nodes, crs="EPSG:4326")
nodes = nodes.to_crs("EPSG:25833")
nodes
```

```python
NL = nodes[nodes["Node"] == "Netherlands"]["geometry"].reset_index()
BE = nodes[nodes["Node"] == "Belgium"]["geometry"].reset_index()
dist_NL_BE = float(NL.distance(BE)/1000) # km
dist_NL_BE
```

```python
scaling_factor = 175/dist_NL_BE
scaling_factor
```

```python
df_dist_offshore_areas = pd.read_csv("Output/dist_offshore_areas.csv")
df_dist_offshore_areas["distance"] = df_dist_offshore_areas["distance"] * scaling_factor
df_dist_offshore_areas
```

```python
df_offshore_to_onshore_nodes = pd.read_csv("Output/offshore_to_onshore_nodes.csv")
df_offshore_to_onshore_nodes["dist"] = df_offshore_to_onshore_nodes["dist"] * scaling_factor
df_offshore_to_onshore_nodes
```

```python
df_shore_to_main_transmission = pd.read_csv("Output/shore_to_main_transmission.csv")
df_shore_to_main_transmission["dist"] = df_shore_to_main_transmission["dist"] * scaling_factor
df_shore_to_main_transmission
```

Export CSV

```python
df_dist_offshore_areas.to_csv("Output/dist_offshore_areas_scaled.csv", index=False, encoding="utf-8-sig")
df_offshore_to_onshore_nodes.to_csv("Output/offshore_to_onshore_nodes_scaled.csv", index=False, encoding="utf-8-sig")
df_shore_to_main_transmission.to_csv("Output/shore_to_main_transmission_scaled.csv", index=False, encoding="utf-8-sig")
```
