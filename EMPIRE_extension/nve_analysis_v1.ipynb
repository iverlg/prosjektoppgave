### Capacity of wind park (base, low and high case)
Source: NVE Oversiktskart - Produksjonsareal_Sirkel

```python
CAPACITY_DENSITY_BASE = 5 #MW/km2
CAPACITY_DENSITY_LOW = 3.5 #MW/km2
CAPACITY_DENSITY_HIGH = 7.5 #MW/km2
```

```python
df["max_capacity_base"] = df["omradeAreal_km2"] * CAPACITY_DENSITY_BASE
df["max_capacity_low"] = df["omradeAreal_km2"] * CAPACITY_DENSITY_LOW
df["max_capacity_high"] = df["omradeAreal_km2"] * CAPACITY_DENSITY_HIGH
df
```

### Calculate distance between areas

```python
import pandas as pd

gdf = df[["omradeNavn", "geometry"]]

result_gdf = None
for row in gdf.itertuples():
    other_gdf = gdf[gdf.index < row.Index]
    distances = other_gdf.geometry.distance(row.geometry)
    distances_gdf = other_gdf.copy()
    distances_gdf["distance"] = distances
    for name, value in row._asdict().items():
        distances_gdf[f"to_{name}"] = value

    if result_gdf is None:
        result_gdf = distances_gdf
    else:
        result_gdf = pd.concat([result_gdf, distances_gdf])

result_gdf
```

```python
distances = result_gdf[["omradeNavn", "to_omradeNavn", "distance"]]
distances["distance_km"] = distances["distance"] / 1000
distances
```

### Approx distance from NVE parks to other parks in EMPIRE model

Mapping done based on data set from: https://www.wins50.nl/data/

EMPIRE areas use grouping of these projects, so one (rather arbitrary; closest project to NO) of the projects are chosen as proxy for distance measures (to be used for calculating transmission cable cost/km)

```python
EU_parks = gpd.read_file("Data/wins50_windfarm_scenarios.dbf")
EU_parks.crs = "EPSG:4326"
EU_parks
```

Mapping from project to EMPIRE node

```python
PROJECT_TO_EMPIRE_NODE = dict({
    "Moray East (A)": "Moray Firth",
    "Berwick Bank": "Firth of Forth",
    "Dogger Bank C": "Dogger Bank",
    "Hornsea 1": "Hornsea",
    "Outer Dowsing": "Outer Dowsing",
    "Norfolk Boreas": "Norfolk",
    "East Anglia Hub - One North": "East Anglia",
    "Borssele 1": "Borssele",
    "Hollandse Kust Noordwest": "Hollandsee Kust",
    "Global Tech I": "Helgoländer Bucht",
    "Future_075": "Nordsøen",
})

PROJECT_FILTER_LIST = PROJECT_TO_EMPIRE_NODE.keys()
```

```python
EU_parks_proxy = EU_parks[EU_parks["name"].isin(PROJECT_FILTER_LIST)]
EU_parks_proxy = EU_parks_proxy.drop_duplicates(subset=["name"])
EU_parks_proxy = EU_parks_proxy[["name", "geometry"]]
EU_parks_proxy["name"] = EU_parks_proxy["name"].apply(lambda n: PROJECT_TO_EMPIRE_NODE.get(n))
EU_parks_proxy
```

```python
NO_parks = df.to_crs("EPSG:4326")
NO_parks = NO_parks[["omradeNavn", "geometry"]].rename(columns={"omradeNavn": "name"})
NO_parks
```

```python
combined = pd.concat([EU_parks_proxy, NO_parks], ignore_index=True)
combined = combined.to_crs("EPSG:25833") # Same as for NO, probably not 100% accurate, but approx.
combined
```

```python
result_combined = None
for row in combined.itertuples():
    other_combined = combined[combined.index < row.Index]
    distances = other_combined.geometry.distance(row.geometry)
    distances_combined = other_combined.copy()
    distances_combined["distance"] = distances
    for name, value in row._asdict().items():
        distances_combined[f"to_{name}"] = value

    if result_combined is None:
        result_combined = distances_combined
    else:
        result_combined = pd.concat([result_combined, distances_combined])

result_combined
```

```python
EU_distances = result_combined[["name", "to_name", "distance"]]
EU_distances["distance_km"] = EU_distances["distance"] / 1000
EU_distances
```

Data set to only add newly calculated values from NVE areas to other areas; keep distance between others as in EMPIRE already

```python
NO_to_EU_distances = EU_distances
EU_AREAS = PROJECT_TO_EMPIRE_NODE.values()
NO_to_EU_distances = NO_to_EU_distances[~(NO_to_EU_distances["name"].isin(EU_AREAS) & NO_to_EU_distances["to_name"].isin(EU_AREAS))]
NO_to_EU_distances
```

### Handling of floating/grounded technology type

Sønnavind A more likely only floating; Sørvest A, D & E most likely grounded but floating possible [both incl] (source: NVE details)

```python
NO_parks_tech = df 
NO_parks_tech.loc[NO_parks_tech["omradeNavn"] == "Sønnavind A", "anleggType"] = "Flytende"
NO_parks_tech
```

```python
df_both_techs = NO_parks_tech[NO_parks_tech["anleggType"] == "Flytende og bunnfast"]
both_techs_parks = list(df_both_techs["omradeNavn"])

df_floating_tech = df_both_techs.copy()
df_grounded_tech = df_both_techs.copy()
df_floating_tech["anleggType"] = "Flytende"
df_grounded_tech["anleggType"] = "Bunnfast"
df_both_techs = pd.concat([df_floating_tech, df_grounded_tech])
df_both_techs
```

```python
NO_parks_tech = NO_parks_tech[~NO_parks_tech["omradeNavn"].isin(both_techs_parks)]
NO_parks_tech = pd.concat([NO_parks_tech, df_both_techs])
NO_parks_tech = NO_parks_tech[["omradeNavn", "anleggType", "nordpool_area", "omradeAreal_km2"]]
NO_parks_tech
```
