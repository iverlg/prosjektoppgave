```python
import pandas as pd 
from constants import NO_NODES, ENERGY_HUBS

NO_NODES_INCL_ONSHORE = NO_NODES + [f"NO{i}" for i in range(1, 6)]
ENERGY_HUBS = [eh.replace(" ", "") for eh in ENERGY_HUBS]

CASE = "NOgrid" # "NOgrid"

df = pd.read_csv(f"Results/{CASE}/results_output_transmission.csv")
df = df[(df["BetweenNode"].isin(NO_NODES)) | (df["BetweenNode"].isin(NO_NODES))]
df["DiscountedInvestmentCost_Euro"] = df["DiscountedInvestmentCost_EuroPerMW"] # Not EUR per MW but EUR
df
```

```python
total_hub_cost_BEUR = 31744.2/1000 if CASE == "Base" else 0 # From total_cost_NS calcs
NO_share_of_hub = 20/187.5 if CASE == "Base" else 0 # In base case
```

```python
# Assume 50/50 split among different countries
cost_split_diff_countries = 0.5
total_cables_cost_EUR = 0

for _, df_row in df.iterrows():
    if df_row["BetweenNode"] in NO_NODES_INCL_ONSHORE and df_row["AndNode"] in NO_NODES_INCL_ONSHORE:
        total_cables_cost_EUR += df_row["DiscountedInvestmentCost_Euro"]
    elif df_row["BetweenNode"] in ENERGY_HUBS or df_row["AndNode"] in ENERGY_HUBS:
        total_cables_cost_EUR += df_row["DiscountedInvestmentCost_Euro"] * NO_share_of_hub
    else:
        total_cables_cost_EUR += df_row["DiscountedInvestmentCost_Euro"] * cost_split_diff_countries
```

```python
NO_cost_of_hubs_base_BEUR = round(total_hub_cost_BEUR * NO_share_of_hub, 1)
print(f"Total NO cost for hubs: {NO_cost_of_hubs_base_BEUR} BEUR")
NO_cost_of_hubs_per_year_base_BEUR = round(NO_cost_of_hubs_base_BEUR / 40, 2)
print(f"Average cables cost per year for NCS OWPs: {NO_cost_of_hubs_per_year_base_BEUR} BEUR\n")

total_cables_cost_BEUR = round(total_cables_cost_EUR/10**9,1)
print(f"Total cables cost for NCS OWPs: {total_cables_cost_BEUR} BEUR")
total_cables_cost_per_year_BEUR = round(total_cables_cost_BEUR / 40, 2)
print(f"Average cables cost per year for NCS OWPs: {total_cables_cost_per_year_BEUR} BEUR\n")

total_costs_incl_hubs_BEUR = NO_cost_of_hubs_base_BEUR + total_cables_cost_BEUR
total_costs_per_year_incl_hubs_BEUR = round(NO_cost_of_hubs_per_year_base_BEUR + total_cables_cost_per_year_BEUR, 2)

print(f"Total costs incl hubs for NCS OWPs: {total_costs_incl_hubs_BEUR} BEUR")
print(f"Average total costs incl hubs per year for NCS OWPs: {total_costs_per_year_incl_hubs_BEUR} BEUR")
```
