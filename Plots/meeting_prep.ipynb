#### Cable capacity to be installed by 2050

```python
import pandas as pd

df_trans = pd.read_csv("base/results_output_transmission.csv")
df_trans
```

Define offshore and NO nodes

```python
OFFSHORE_NODES = ["Moray Firth","Firth of Forth","Dogger Bank",
                     "Hornsea","Outer Dowsing","Norfolk","East Anglia",
                     "Borssele","Hollandsee Kust","Helgoländer Bucht",
                     "Nordsøen",'Nordvest A', 'Nordvest C', 'Vestavind A', 
                     'Sønnavind A', 'Sørvest C', 'Nordvest B', 'Vestavind F', 
                     'Sørvest E', 'Sørvest A', 'Vestavind B', 'Vestavind C',
                     'Vestavind D', 'Sørvest F', 'Sørvest B', 'Nordavind B', 
                     'Nordavind A', 'Nordavind D', 'Nordavind C', 'Vestavind E', 'Sørvest D', 
                     "Energyhub EU", "Energyhub North", "Energyhub Central"]
OFFSHORE_NODES = [node.replace(" ", "") for node in OFFSHORE_NODES]

OFFSHORE_NODES_NO = ['Nordvest A', 'Nordvest C', 'Vestavind A', 
                     'Sønnavind A', 'Sørvest C', 'Nordvest B', 'Vestavind F', 
                     'Sørvest E', 'Sørvest A', 'Vestavind B', 'Vestavind C',
                     'Vestavind D', 'Sørvest F', 'Sørvest B', 'Nordavind B', 
                     'Nordavind A', 'Nordavind D', 'Nordavind C', 'Vestavind E', 'Sørvest D',
                     "Energyhub EU", "Energyhub North", "Energyhub Central"]
OFFSHORE_NODES_NO = [node.replace(" ", "") for node in OFFSHORE_NODES_NO]

NO_NODES = [f"NO{i}" for i in range(1, 6)]

NO_NODES
```

Define new tables; 
1) All offshore connections
2) Offshore NO-connections (only NO offshore areas incl. hubs to look at offshore cables)

```python
#1
df_trans_offshore = df_trans[(df_trans["BetweenNode"].isin(OFFSHORE_NODES)) | (df_trans["AndNode"].isin(OFFSHORE_NODES))]\
                        .reset_index(drop=True)
df_trans_offshore
```

```python
#2
df_trans_NO = df_trans[((df_trans["BetweenNode"].isin(OFFSHORE_NODES_NO)) & (df_trans["AndNode"].isin(OFFSHORE_NODES_NO))) | \
                       ((df_trans["BetweenNode"].isin(NO_NODES)) & (df_trans["AndNode"].isin(OFFSHORE_NODES_NO))) | \
                        ((df_trans["BetweenNode"].isin(OFFSHORE_NODES_NO)) & (df_trans["AndNode"].isin(NO_NODES)))]\
                        .reset_index(drop=True)
df_trans_NO
```

#### Initial transmission capacity (2023)

```python
# First period updated with current installed capacity, threshold installed cap > 1 MW
df_trans_offshore_2023 = df_trans_offshore[(df_trans_offshore["Period"] == "2020-2025") & \
                                           (df_trans_offshore["transmissionInstalledCap_MW"] > 1)].reset_index(drop=True)
df_trans_NO_2023 = df_trans_NO[(df_trans_NO["Period"] == "2020-2025") & \
                                           (df_trans_NO["transmissionInstalledCap_MW"] > 1)].reset_index(drop=True)

initial_cap_trans_offshore_2023 = df_trans_offshore_2023["transmissionInstalledCap_MW"].sum() \
                                    - df_trans_offshore_2023["transmissionInvCap_MW"].sum()
initial_cap_trans_NO_2023 = df_trans_NO_2023["transmissionInstalledCap_MW"].sum() \
                                    - df_trans_NO_2023["transmissionInvCap_MW"].sum()

print(f"Initial trans capacity 2023 all offshore areas: {round(initial_cap_trans_offshore_2023 / 1000, 1)} GW")
print(f"Initial trans capacity 2023 NO areas: {round(initial_cap_trans_NO_2023 / 1000, 1)} GW")
```

##### Transmission capacity installed in 2050

```python
# First period updated with current installed capacity, threshold installed cap > 1 MW
df_trans_offshore_2050 = df_trans_offshore[(df_trans_offshore["Period"] == "2050-2055") & \
                                           (df_trans_offshore["transmissionInstalledCap_MW"] > 1)].reset_index(drop=True)
df_trans_NO_2050 = df_trans_NO[(df_trans_NO["Period"] == "2050-2055") & \
                                           (df_trans_NO["transmissionInstalledCap_MW"] > 1)].reset_index(drop=True)

cap_trans_offshore_2050 = df_trans_offshore_2050["transmissionInstalledCap_MW"].sum() \
                                    - df_trans_offshore_2050["transmissionInvCap_MW"].sum()
cap_trans_NO_2050 = df_trans_NO_2050["transmissionInstalledCap_MW"].sum() \
                                    - df_trans_NO_2050["transmissionInvCap_MW"].sum()

print(f"Initial trans capacity 2050 all offshore areas: {round(cap_trans_offshore_2050 / 1000, 1)} GW")
print(f"Initial trans capacity 2050 NO areas: {round(cap_trans_NO_2050 / 1000, 1)} GW\n")

# New cable capacity needed
print(f"New cable trans capacity needed by 2050 all offshore areas: {round((cap_trans_offshore_2050-initial_cap_trans_offshore_2023) / 1000, 1)} GW")
print(f"New cable trans capacity needed by 2050 NO areas: {round((cap_trans_NO_2050-initial_cap_trans_NO_2023) / 1000, 1)} GW")
```

#### Compare to production capacity

```python
df_prod = pd.read_csv("base/results_output_gen.csv")
df_prod
```

```python
#1
df_prod_offshore = df_prod[(df_prod["Node"].isin(OFFSHORE_NODES))].reset_index(drop=True)
df_prod_offshore
```

#### Production capacity installed 2050

```python
# First period updated with current installed capacity, threshold installed cap > 1 MW
df_prod_offshore_2050 = df_prod_offshore[(df_prod_offshore["Period"] == "2050-2055") & \
                                           (df_prod_offshore["genInstalledCap_MW"] > 1)].reset_index(drop=True)
df_prod_NO_2050 = df_prod_NO[(df_prod_NO["Period"] == "2050-2055") & \
                                           (df_prod_NO["genInstalledCap_MW"] > 1)].reset_index(drop=True)

cap_prod_offshore_2050 = df_prod_offshore_2050["genInstalledCap_MW"].sum() \
                                    - df_prod_offshore_2050["genInvCap_MW"].sum()
cap_prod_NO_2050 = df_prod_NO_2050["genInstalledCap_MW"].sum() \
                                    - df_prod_NO_2050["genInvCap_MW"].sum()

print(f"Initial prod capacity 2050 all offshore areas: {round(cap_prod_offshore_2050 / 1000, 1)} GW")
print(f"Initial prod capacity 2050 NO areas: {round(cap_prod_NO_2050 / 1000, 1)} GW\n")

# New cable capacity needed
print(f"New prod capacity needed by 2050 all offshore areas: {round((cap_prod_offshore_2050-initial_cap_prod_offshore_2023) / 1000, 1)} GW")
print(f"New prod capacity needed by 2050 NO areas: {round((cap_prod_NO_2050-initial_cap_prod_NO_2023) / 1000, 1)} GW")
```

#### Additional trans capacity above max prod

```python
add_trans_cap_offshore_2023 = initial_cap_trans_offshore_2023 - initial_cap_prod_offshore_2023
add_trans_cap_NO_2023 = initial_cap_trans_NO_2023 - initial_cap_prod_NO_2023

add_trans_cap_offshore_2050 = cap_trans_offshore_2050 - cap_prod_offshore_2050
add_trans_cap_NO_2050 = cap_trans_NO_2050 - cap_prod_NO_2050

# Additional cap above max prod (abs)
print(f"Additional trans cap above max prod 2023 all offshore areas: {round((add_trans_cap_offshore_2023) / 1000, 1)} GW")
print(f"Additional trans cap above max prod 2023 NO areas: {round((add_trans_cap_NO_2023) / 1000, 1)} GW\n")
print(f"Additional trans cap above max prod 2050 all offshore areas: {round((add_trans_cap_offshore_2050) / 1000, 1)} GW")
print(f"Additional trans cap above max prod 2050 NO areas: {round((add_trans_cap_NO_2050) / 1000, 1)} GW\n")

# Additional cap above max prod (%)
print(f"Additional trans cap above max prod 2023 all offshore areas: {round(add_trans_cap_offshore_2023/initial_cap_prod_offshore_2023*100, 1)}%")
print(f"Additional trans cap above max prod 2023 NO areas: 0.0%\n") # Manually punched 0% since initial_cap_prod_NO_2023 = 0 (DIV0-error)
print(f"Additional trans cap above max prod 2050 all offshore areas: {round(add_trans_cap_offshore_2050/cap_prod_offshore_2050*100, 1)}%")
print(f"Additional trans cap above max prod 2050 NO areas: {round(add_trans_cap_NO_2050/cap_prod_NO_2050*100, 1)}%")
```

### Curtailment of offshore wind areas

In this work, curtailment is defined as the unused available power from intermittent renewable sources.

```python
df_curt = pd.read_csv("base/results_output_curtailed_prod.csv")
df_curt
```

```python
#1
df_curt_offshore = df_curt[(df_curt["Node"].isin(OFFSHORE_NODES))].reset_index(drop=True)
df_curt_offshore
```

```python
#2
df_curt_NO = df_curt[(df_curt["Node"].isin(OFFSHORE_NODES_NO))].reset_index(drop=True)
df_curt_NO
```

#### Initial curtailment (2023)

```python
# First period, threshold expected annual power > 0 GWh
df_curt_offshore_2023 = df_curt_offshore[(df_curt_offshore["Period"] == "2020-2025") & \
                                           (df_curt_offshore["Expected total available power_GWh"] > 0)].reset_index(drop=True)
df_curt_NO_2023 = df_curt_NO[(df_curt_NO["Period"] == "2020-2025") & \
                                           (df_curt_NO["Expected total available power_GWh"] > 0)].reset_index(drop=True)

avg_curt_offshore_2023 = df_curt_offshore_2023["Expected annual curtailment ratio of total capacity_%"].mean() 
avg_curt_NO_2023 = df_curt_NO_2023["Expected annual curtailment ratio of total capacity_%"].mean() 

print(f"Average curtailment 2023 all offshore areas: {round(avg_curt_offshore_2023, 1)}%")
print(f"Average curtailment 2023 NO areas: {round(avg_curt_NO_2023, 1)}%")
```

#### Curtailment 2050

```python
# 2050, threshold expected annual power > 0 GWh
df_curt_offshore_2050 = df_curt_offshore[(df_curt_offshore["Period"] == "2050-2055") & \
                                           (df_curt_offshore["Expected total available power_GWh"] > 0)].reset_index(drop=True)
df_curt_NO_2050 = df_curt_NO[(df_curt_NO["Period"] == "2050-2055") & \
                                           (df_curt_NO["Expected total available power_GWh"] > 0)].reset_index(drop=True)

avg_curt_offshore_2050 = df_curt_offshore_2050["Expected annual curtailment ratio of total capacity_%"].mean() 
avg_curt_NO_2050 = df_curt_NO_2050["Expected annual curtailment ratio of total capacity_%"].mean() 

print(f"Average curtailment 2050 all offshore areas: {round(avg_curt_offshore_2050, 1)}%")
print(f"Average curtailment 2050 NO areas: {round(avg_curt_NO_2050, 1)}%")
```
