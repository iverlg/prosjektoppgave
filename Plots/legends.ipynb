```python
import matplotlib.pyplot as plt
from constants import (HUB_TO_COLOR, NO_NODES, TECH_TO_COLOR,
                       WIND_FARM_NODES_TO_COLOR)
from matplotlib.patches import Patch
```

### For total generation by source plots

```python
legend_elements = [Patch(facecolor=color, label=tech) for (tech, color) in TECH_TO_COLOR.items()]

fig, ax = plt.subplots(figsize=(20,1))
ax.legend(handles=legend_elements, loc='center', ncols=6, frameon=False, fontsize=14)
plt.axis("off")
plt.show()
```

Fix incl techs after looking at actual results to avoid techs not included!!

```python
INCL_TECHS = ['Windonshore', 'Solar', 'Bio', 'Nuclear', 'Windoffshoregrounded', 'GasOCGT', 'Hydrorun-of-the-river', 'Hydroregulated', 'GasCCGT', 'Windoffshorefloating', 'Lignite', 'Bio10cofiring', 'Waste', 'Wave', 'Geo', 'Bioexisting']

legend_elements = [Patch(facecolor=TECH_TO_COLOR[tech], label=tech) for tech in INCL_TECHS]
fig, ax = plt.subplots(figsize=(20,0.5))
ax.legend(handles=legend_elements, loc='center', ncols=6, frameon=False, fontsize=14)
plt.axis("off")
plt.savefig("SavedFigs/Legends/EnergyMix_legend", bbox_inches="tight")
plt.show()
```
