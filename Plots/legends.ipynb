#### Sensitivity legend

```python
CASE_ORDER = ['-30%', '-20%', '-10%', 'Base', '+10%', '+20%','+30%']
COLORS = ['#2166ac', '#6a86b7', '#9da8c2', '#cccccc', '#cd9792', '#c4615c', '#b2182b']

CASE_ORDER_TO_COLOR = dict({})

for i in range(len(CASE_ORDER)):
    CASE_ORDER_TO_COLOR[CASE_ORDER[i]] = COLORS[i]

legend_elements = [Patch(facecolor=CASE_ORDER_TO_COLOR[case], label=case) for case in CASE_ORDER]
fig, ax = plt.subplots(figsize=(20,0.5))
ax.legend(handles=legend_elements, loc='center', ncols=7, frameon=False, fontsize=20)
plt.axis("off")
plt.savefig("SavedFigs/Legends/Sensitivity_legend", bbox_inches="tight")
plt.show()
```
